## **2. SIMPLE EXAMPLES**

### MATPLOTLIB
**Here's what we can do using matplotlib:**
* **Basic**: Basic plot types, usually y versus x.
* **Plots of arrays and fields**: Plotting for arrays of data Z(x, y) and fields U(x, y), V(x, y).
* **Statistics plots**: Plots for statistical analysis.
* **Unstructured coordinates** : Sometimes we collect data z at coordinates (x,y) and want to visualize as a contour. Instead of gridding the data and then using contour, we can use a triangulation algorithm and fill the triangles.
* **3D** : 3D plots using the mpl_toolkits.mplot3d library.

*Reminder:* At this step, these codes may seen useless but these are good for getting into the swing of visualization.  

```python
# PLOT

import matplotlib.pyplot as plt
import numpy as np

# make data
x = np.linspace(0, 20, 1000)
y = 4 + 3 * np.sin(2 * x)

# plot
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 8), xticks=np.arange(1, 10),
       ylim=(0, 8), yticks=np.arange(1, 10))

plt.grid(True)
plt.show()
```

```python
# BAR

# make data:
np.random.seed(3)
x = 0.5 + np.arange(20)
y = np.random.uniform(2, 8, len(x))

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="red", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

print("x:", x,"\n", "y:",y)
plt.grid(True)
plt.show()
```

```python
#SCATTER

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 36)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(16, 100, len(x))
colors = np.random.uniform(16, 100, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=-50, vmax=50)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.grid(True)
plt.show()
```

```python
#HISTOGRAM

# make data
np.random.seed(1)
x = x = 4 + np.random.normal(0, 2, 200)

# plot:
fig, ax = plt.subplots()

ax.hist(x, bins=12, color= "yellow", linewidth=0.7, edgecolor="red")

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.grid(True)
plt.show()
```

```python
# PIE CHART

# make data
x = [1, 2, 3, 4, 5]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.8, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=4, center=(5, 5),
       wedgeprops={"linewidth": 2, "edgecolor": "white"}, frame=True)

plt.show()
```

```python
# 2D HISTOGRAM : NOISE 

# make data: correlated + noise
np.random.seed(1)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3

# plot:
fig, ax = plt.subplots()

ax.hist2d(x, y, bins=(np.arange(-3, 3, 0.1), np.arange(-3, 3, 0.1)))
ax.set(xlim=(-2, 2), ylim=(-3, 3))

plt.grid()
plt.show()
```

```python
#3 SURFACE
from matplotlib import cm

# Make data
X = np.arange(-100, 100, 20)
Y = np.arange(-100, 100, 20)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min()* 3 , cmap=cm.Blues)

ax.set(xticklabels=["x"],
       yticklabels=["y"],
       zticklabels=["z"])

plt.show()
```

### SEABORN

```python
import seaborn as sns
from matplotlib import rcParams
sns.set_theme(style="darkgrid")

# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")

# Plot the responses for different events and regions
p = sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",markers=True,
             data=fmri)

plt.subplots(figsize = (5,5))
```

```python
#Grouped boxplot

sns.set_theme(style="ticks", palette="pastel")

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["m", "y"],
            data=tips)
sns.despine(offset=10, trim=True)
```

```python
#Annotated heatmaps
sns.set_theme()

# Load the example flights dataset and convert to long-form
flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
```

```python
#Scatterplot with varying point sizes and hues
sns.set_theme(style="white")

# Load the example mpg dataset
mpg = sns.load_dataset("mpg")

# Plot miles per gallon against horsepower with other semantics
sns.relplot(x="horsepower", y="mpg", hue="origin", size="weight",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=mpg)
plt.grid()
```

```python
#Multiple Linear Regression

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Plot sepal width as a function of sepal_length across days
g = sns.lmplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="species",
    height=5
)

# Use more informative axis labels than are provided by default
g.set_axis_labels("Snoot length (mm)", "Snoot depth (mm)")
```

```python
i = "Forest Area"
[int(s) for s in i.split() if s.isdigit()]
```

```python
droped_years = []
for i in data.columns:
    year = [int(s) for s in i.split() if s.isdigit()]
    if year != []:
        year = [int(s) for s in i.split() if s.isdigit()][0]
        if (year % 5) != 0:
            droped_years.append(i)
        
        
print(droped_years)    
```

```python
df = data.drop(droped_years, axis=1)
df.head()
```

```python
df.columns
```

### VISUALIZATION

```python
# data visualization libraries 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#Comparison of forest areas by continents in 2020: BOXPLOT

sns.boxplot(x="Forest Area 2020", y="Continent", data=df,
            whis=[0, 100], width=.6, palette="vlag")
# Add in points to show each observation
sns.stripplot(x="Forest Area 2020", y="Continent", data=df,
              size=4, color=".3", linewidth=0)
```

```python
#Comparison by continent, how has forest area changed over the years?
years = ['Forest Area 1990','Forest Area 1995', 'Forest Area 2000',
         'Forest Area 2005','Forest Area 2010', 'Forest Area 2015', 'Forest Area 2020']
for i in years:
    sns.boxplot(x=i, y="Continent", data=df,
                whis=[0, 100], width=.6, palette="vlag")
    
    sns.stripplot(x=i, y="Continent", data=df,
              size=4, color=".3", linewidth=0)
    plt.show()
    
```
